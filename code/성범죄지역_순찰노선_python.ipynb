### 4. 순찰노선 그리기
- 경찰서 인근을 시작으로 최적의 순찰노선 그리기
- PolyLine()으로 루트 시각화

```python

line_route = [(37.456714, 127.127596), #최벅람 
              (37.454939, 127.130419), #변대홉 
              (37.454470, 127.130988), #권몁혈 
              (37.453925, 127.131642), #황변휴
              (37.454334, 127.129357), #곽독평 
              (37.454930, 127.129582), #송졍석 
              (37.455288, 127.129164), #이볍구 
              (37.455191, 127.128437), #홍길동 
              (37.455022, 127.127794), #박념동 
              (37.455744, 127.127722)] #김겹독   


folium.PolyLine(locations=line_route).add_to(m)

#folium.PolyLine(locations=line_coords, # 선의 위경도 좌표
                #color='blue',  # 선 색상
                #weight=5,  # 선 굵기
                #opacity=0.7,  # 선 투명도
                #tooltip="memo").add_to(m)
m
```
